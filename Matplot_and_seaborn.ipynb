```python
import seaborn as sns
import matplotlib.pyplot as plt

# Given Data
tips_data = [
    (16.99, 1.01, 'Female','Mon', 'Dinner', 2),
    (10.34, 1.66, 'Male','Mon', 'Dinner', 3),
    (21.01, 3.5, 'Male', 'Tue', 'Dinner', 3),
    (23.68, 3.31, 'Male', 'Tue', 'Dinner', 2),
    (24.59, 3.61, 'Female','Tue', 'Dinner', 4),
    (25.29, 4.71, 'Male', 'Wed', 'Dinner', 4),
    (8.77, 2.0, 'Male','Thu', 'Dinner', 2),
    (26.88, 3.12, 'Male','Thu', 'Dinner', 4),
    (15.04, 1.96, 'Male','Sun', 'Dinner', 2),
    (14.78, 3.23, 'Male','Sun', 'Dinner', 2)
]
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the built-in 'iris' dataset from Seaborn
iris = sns.load_dataset("iris")

# Create a pair plot of the 'iris' dataset
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(10,10))
for i in range(4):
    for j in range(4):
        ax = axes[i,j]
        if i == j:
            ax.hist(iris.iloc[:, i], bins=15)
        else:
            ax.scatter(iris.iloc[:, j], iris.iloc[:, i], s=30, edgecolor='k')
        if i == 0:
            ax.set_title(iris.columns[j])
        if j == 0:
            ax.set_ylabel(iris.columns[i])

# Create a heatmap of the correlations between variables in the 'iris' dataset
corr = iris.corr()
fig, ax = plt.subplots(figsize=(8, 8))
im = ax.imshow(corr.values, cmap='coolwarm')
ax.set_xticks(range(len(corr.columns)))
ax.set_yticks(range(len(corr.columns)))
ax.set_xticklabels(corr.columns)
ax.set_yticklabels(corr.columns)
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")
for i in range(len(corr.columns)):
    for j in range(len(corr.columns)):
        text = ax.text(j, i, '{:.2f}'.format(corr.values[i, j]),
                       ha="center", va="center", color="black")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips") #read_csv

smokers = tips[tips['smoker'] == 'Yes']
non_smokers = tips[tips['smoker'] == 'No']

# Create a figure and axis object
fig, (ax1,ax2) = plt.subplots(1,2)

# Create two scatter plots, one for smokers and one for non-smokers
ax1.scatter(smokers['total_bill'], smokers['tip'], color='orange', label='Yes')
ax2.scatter(non_smokers['total_bill'], non_smokers['tip'], color='blue', label='No')
plt.rcParams.update({'figure.figsize': (5,5), 'font.size': 5})

# Set the chart title and axis labels
ax1.set_title('Tips by Total Bill and Smoker Status')
ax1.set_xlabel('Total Bill')
ax1.set_ylabel('Tip')
ax2.set_title('Tips by Total Bill and Smoker Status')
ax2.set_xlabel('Total Bill')
ax2.set_ylabel('Tip')
# Add a legend
ax1.legend()
ax2.legend()
# Display the plot
plt.show()
```
